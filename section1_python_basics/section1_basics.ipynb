## 2. Assignment 1 — Grade from degree

```python
def assignment_1(name, department, degree):
    if   degree >= 85: grade = "Excellent"
    elif degree >= 75: grade = "Very Good"
    elif degree >= 65: grade = "Good"
    elif degree >= 50: grade = "Pass"
    else:              grade = "Fail"
    return f"Hello {name}, your department is {department} and your degree is {grade}!"

# Demo with hard-coded inputs (replace with input() if you want an interactive run)
print(assignment_1("Mohamed", "IS", 92))
print(assignment_1("Ahmed",   "IS", 78))
print(assignment_1("Yasmin",  "IS", 67))
print(assignment_1("Karim",   "IS", 54))
print(assignment_1("Omar",    "IS", 38))
```

## 3. Assignment 2 — BMI calculator
A natural fit for FitZone Sports — many customers ask about which sports
equipment matches their fitness profile, and BMI is the first input the
in-store coach asks for.

```python
def bmi_status(name, weight_kg, height_cm):
    height_m = height_cm / 100
    bmi = weight_kg / (height_m * height_m)
    if   bmi <= 18.5:                  status = "UNDERWEIGHT"
    elif 18.5 < bmi <= 24.9:           status = "NORMAL WEIGHT"
    elif 24.9 < bmi <= 29.9:           status = "OVERWEIGHT"
    else:                              status = "OBESE"
    return f"{name}, your BMI value is {bmi:.2f} and you are: {status}"

print(bmi_status("Mohamed", 70,  175))
print(bmi_status("Ahmed",   85,  178))
print(bmi_status("Yasmin",  55,  165))
print(bmi_status("Karim",   95,  180))
```

## 4. Assignment 3 — DataFrame operations

```python
import pandas as pd

df = pd.DataFrame({
    "col1": [1, 2, 3, 4, 7],
    "col2": [4, 5, 6, 9, 5],
    "col3": [7, 8, 12, 1, 11],
})
df
```

```python
# (a) number of rows & columns
print("rows :", df.shape[0])
print("cols :", df.shape[1])
```

```python
# (b) random 50% rows
df.sample(frac=0.5, random_state=1)
```

```python
# (c) sort by col1 DESC
df.sort_values('col1', ascending=False)
```

```python
# (d) filter where col2 >= 7
df.query('col2 >= 7')
```

```python
# (e) oldest 3 rows = the last 3 rows of the frame
df.tail(3)
```
